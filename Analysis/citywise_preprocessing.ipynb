```python
import pandas as pd

dataset = "datasets/hyderabad.csv"
data = pd.read_csv(dataset)
df=pd.DataFrame(data)
df.tail()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

### Selecting only required columns

```python
df_new = df.iloc[:,[0,1,2,10,11,18,20,23,24]]
df_new.head()
```

### Calulating the average temparature using the max and min temperature

```python
df_new['averagetempC'] = (df_new.iloc[:,1] + df_new.iloc[:,2])/2
df_new = df_new.drop(['maxtempC','mintempC'],axis=1)
df_new.head()
```

### Calculating duration of sunlight in minutes from sunrise and sunset timings

```python
from datetime import datetime

def calc_sunlight_duration(row):
    sunrise = datetime.strptime(row['sunrise'], '%I:%M %p')
    sunset = datetime.strptime(row['sunset'], '%I:%M %p')
    duration = (sunset - sunrise).total_seconds() / 60
    return duration

df_new['SunLightDuration'] = df_new.apply(calc_sunlight_duration, axis=1)
df_new
```

```python
df_new = df_new.drop(['sunrise','sunset'],axis=1)
df_new
```

### Extracting 7 years of data 


```python
df_new['date_time'] = pd.to_datetime(df_new['date_time'])
start_date = pd.to_datetime('2013-01-01') 
end_date = pd.to_datetime('2020-01-01') 
df_filtered = df_new[(df_new['date_time'] >= start_date) & (df_new['date_time'] <= end_date)]
df_filtered
```

### Saving into final dataframe to convert to csv format

```python
df_final = pd.DataFrame({})
df_final["date_time"]=df_filtered["date_time"]
df_final["Temperature"] = df_filtered["averagetempC"]
df_final["Pressure"] = df_filtered["pressure"] - 1000
df_final["Humidity"] = df_filtered["humidity"]
df_final["WindDirection(Degrees)"] = df_filtered["winddirDegree"]
df_final["Speed"] = df_filtered["windspeedKmph"]
df_final["SunLightDuration"] = df_filtered["SunLightDuration"]
df_final
```

### Prediction of solar radiation using random forest

```python
from sklearn.ensemble import RandomForestRegressor


df1 = pd.read_csv('datasets/SolarPrediction_processed.csv')

model = RandomForestRegressor(n_estimators=100, random_state=42)

X_train = df1.drop(['SolarRadiation'], axis=1)
y_train = df1['SolarRadiation']
model.fit(X_train, y_train)


X_test = df_final.drop(['date_time'], axis=1)
y_pred = model.predict(X_test)

df_final['SolarRadiation'] = y_pred

df_final
```

### Calculating energy generated

```python
# Energy_generated_per_hour = area_of_panel * efficiency_of_panel * solar_radiation
area_of_panel = 1
efficiency = 0.7
df_final["EnergyGenerated"] = ( df_final["SolarRadiation"] * area_of_panel * efficiency)
df_final 
```
